### Finding Least squares line

```python
n=1248
xsum=sum(data.ENGINE_SIZE)
ysum=sum(data.FUEL_CONSUMPTION)
xy=sum(data.ENGINE_SIZE*data.FUEL_CONSUMPTION)
xsqsum=sum(data.ENGINE_SIZE**2)
ysqsum=sum(data.FUEL_CONSUMPTION)
m=(n*xy-xsum*ysum)/(n*xsqsum-xsum**2)
b=(ysum-m*(xsum))/n
x=data.ENGINE_SIZE
y=data.ENGINE_SIZE*m+b
print(m,b)
```

### Correlation coeff

```python
Corr=(n*xy-xsum*ysum)/(math.sqrt(n*xsqsum-xsum**2)*(n*ysqsum-ysum**2)**0.5)
print('{:.10f}'.format(Corr))
```
