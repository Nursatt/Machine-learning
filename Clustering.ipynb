# Этот код загружает набор данных Iris, инициализирует модель кластеризации k-средних с тремя кластерами, подгоняет модель к данным и выводит результаты на график. На полученном рисунке должны быть видны три отдельные группы цветков ириса, основанные на их длине и ширине чашелистиков.

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
# Load the Iris dataset
iris = load_iris()

# Initialize a k-means clustering model with 3 clusters
kmeans = KMeans(n_clusters=3)

# Fit the model to the data
kmeans.fit(iris.data)

# Get the cluster assignments for each data point
labels = kmeans.predict(iris.data)

# Plot the results
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=labels)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
# Elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(iris.data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Silhouette score
silhouette_avg = silhouette_score(iris.data, labels)
print("The average silhouette_score is :", silhouette_avg)
```
